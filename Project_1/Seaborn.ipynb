```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style ('whitegrid')  ## style of grid lines
titanic = sns.load_dataset('titanic') ## calling dataset 'titanic'
titanic.head() # show head of table
```

```python
sns.jointplot (x = 'fare', y= 'age' , data=titanic)
```

```python
sns.displot(titanic['fare'], bins = 30, color='red', kde = True) 
## kde is for graph line. by standard it is false and not showing # bins = no of grid squres
# it is ploted for one column on x axis and y axis for count or occurence of x axis paraeter
```

```python
sns.boxplot (x = 'class', y= 'age', data =titanic, palette='rainbow', hue = 'sex')
## box plot require x, y and can give hue
```

```python
sns.swarmplot(data =titanic, x = 'class', y = 'age') ## it is mixture of normal and 
```

```python
sns.countplot (x = 'sex', data = titanic, palette='rainbow')
```

```python
# heat map --< need to give correlation file.corr(numeric_only = true) # for heatmap color use cmap
sns.heatmap(titanic.corr(numeric_only= True), cmap= 'coolwarm') 
plt.title ('titanic.corr()')
```

```python
## Heat map editing
sns.heatmap(titanic.corr(numeric_only= True), annot=True, fmt='.1g', vmin= -1 , vmax= 1 , center= 0,cmap='coolwarm',linewidths= 2 , linecolor= 'black',
            cbar_kws= {'orientation' : 'horizontal'}, cbar=True)
```

```python
g = sns.FacetGrid(data=titanic, col = 'sex') ## this line is to create two grids
g.map(plt.hist, 'age')  ## use here matplot histogram function to graph inside above grid
```
