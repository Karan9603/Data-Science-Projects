```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(2,10,5)
y = x**2
print ('x is ',x)
print ('y is',y)
```

```python
##Plot the graph
plt.plot (x,y) # plot the graph
plt.xlabel ('value') # label x axis
plt.ylabel ('values squre') # label y axis
plt.title ('numbers squre') # title of the graph
```

```python
## creating multiplot in same canvas
plt.subplot(1,3,1)  ## subplot is part of plot for which first two digit as dimension and third digit is which plot (position of plot)
plt.plot (x,y, 'red') ## here we are giving value to above selected plot. here plot 1
plt.subplot(1,3,3)  ## here subplot dim is 1*3 matrix and position of subplot is 3
plt.plot (x,x*2, 'Orange') ## here value for subplot position 3. values X axis = x, Y axis = x * 2
```

```python
##another example for subplot with dimension 2*3
plt.subplot(2,3,1)
plt.plot (x,y, 'black')
plt.subplot(2,3,6)
plt.plot (y, y*2, 'blue')
```

```python
## Object Orianted methods ##

fig = plt.figure() ##assign function to vaiable
axes = fig.add_axes([0.1,0.2,0.7,0.6]) ## here we set axis distance in convas and alway in []
axes.plot(x,y) ## we give here x and y axis values
axes.set_xlabel ('number')
axes.set_ylabel ('squre')
axes.set_title ('squre values')
```

```python
## two axis or graph inside graph ## figsize

fig = plt.figure(figsize=(8,6)) ## here figsize function change the size of graph sheet
axes1 = fig.add_axes([0,0,0.8,0.8]) # axis 1 (graph1)
axes2 = fig.add_axes([0.4,0.1, 0.3,0.3]) # axis 2 (graph2)
axes1.plot (y,x)
axes1.set_title ('large graph')
axes2.plot (y, y-2)
axes2.set_title ('small graph')
```

```python
## subplots (matrix graphs) using object oriented
## here dont have add axis location in sheet like before example, system takes automatically. ## we can also use figsize here
fig, axes = plt.subplots(figsize = (8,6)) 
axes.plot (x,y) 
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 3, figsize = (8,6)) 
plt.tight_layout() ## to seperate conjested graphs
axes[0][0].plot (y,x)  ## here index need to give in [] infront of axis eith [][] or [_, _]
axes[1,2].plot (x,y, 'red')
axes [1,2].set_title ('matrix graph')
```

```python
## legend with figure size

fig = plt.figure(figsize=(6,4))
axes = fig.add_axes([0,0,1,1])
axes.plot (x**2, label = 'x squre')
axes.plot (x**3, label = 'x Cube')
axes.legend()  ## here legend fuction used to show two series in one graph with y axis no of count
# axes.legend(loc = 2) here (loc = 2) is for place of color table small chart. currently its upeer left side
plt.xlabel ('numbers')


```

```python
#linewidth
fig = plt.figure()
axes = fig.add_axes ([0,0,1,1])
axes.plot (x,y, color = 'purple', linewidth = 3, alpha = 0.5, linestyle = ':', 
           marker = 'o', markersize = 20, markeredgecolor = 'green', markeredgewidth = 3, markerfacecolor = 'yellow') 
# linewidth chnages ## alpha changes transperancy of line ## linestyle changeds line style like ('--', '-.', ';')
# marker 'o' show the points # markersize show size of marker # markeredgecolor # markeredgewidth # markerfacecolor
```

```python
## crop specific part toi focus on
fig = plt.figure()
axes = fig.add_axes ([0,0,1,1])
axes.plot (x,y, color = 'purple', linewidth = 3, alpha = 0.5, linestyle = ':', marker = 'o') 
axes.set_xlim ([0,8])
axes.set_ylim ([0,60])
```
